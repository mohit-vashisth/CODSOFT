```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("../dataset/titanic_eda.csv")
```

```python
df["is_alone"] = (df["family_size"] == 1).astype(int)
```

```python
df["fare_per_person"] = df["fare"] / df["family_size"]
```

```python
df["age_bin"] = pd.cut(df["age"], bins=[0, 12, 18, 35, 60, 80], labels=["child", "teen", "young", "adult", "senior"])
```

```python
df["class_fare"] = df["pclass"] * df["fare"]
```

```python
df["fare_log"] = np.log1p(df["fare"])  # log(1+fare)
```

```python
df[df["age"] > 80]   # Rare in 1912
df[df["fare"] == 0]  # Possibly crew/staff
```

```python
df.groupby("pclass")["survived"].mean()
df.groupby("sex")["survived"].mean()
df.groupby("title")["survived"].mean()
```

```python
df = pd.get_dummies(df, columns=["age_bin"], prefix="age")
```

```python
df["is_child_female"] = ((df["sex"] == 0) & (df["age"] <= 12)).astype(int)
```

```python
df.to_csv("../dataset/titanic_processed.csv", index=False)
```
