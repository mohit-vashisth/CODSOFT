```python
# ---------------------------------------
# 1. Import Required Libraries
# ---------------------------------------
import pandas as pd
import numpy as np
import joblib
```

```python
# ---------------------------------------
# 2. Load Saved Model and Feature Columns
# ---------------------------------------
model = joblib.load("../models/xgb_movie_rating_model.pkl")
feature_cols = joblib.load("../models/xgb_feature_columns.pkl")

print("‚úÖ Model and feature list loaded successfully")
```

```python
# ---------------------------------------
# 3. Load New Data for Prediction
# ---------------------------------------
# NOTE: New data should have the SAME columns (excluding 'rating', 'votes')
new_data = pd.DataFrame(
    [
        {
            "name": "Example Movie",
            "year_int": 2022,
            "duration": 130.0,
            "genre": "Action,Adventure",  # üîÅ Changed pipe to comma
            "director": "Christopher Nolan",
            "actor 1": "Leonardo DiCaprio",
            "actor 2": "Tom Hardy",
            "actor 3": "Joseph Gordon-Levitt",
            "genre_split": ["Action", "Adventure"],  # üîÅ Changed to list
            "genre_count": 2,
            "director_avg_rating": 8.5,
            "actor_popularity": 8.9,
        }
    ]
)
```

```python
# ---------------------------------------
# 4. Align Columns with Training Data
# ---------------------------------------
# Just to be safe ‚Äî ensure all required columns are present
missing_cols = set(feature_cols) - set(new_data.columns)
if missing_cols:
    raise ValueError(f"‚ùå Missing columns in new data: {missing_cols}")

# Ensure correct column order
new_data = new_data[feature_cols]
```

```python
# ---------------------------------------
# 5. Make Prediction
# ---------------------------------------
predicted_rating = model.predict(new_data)[0]
print(f"üéØ Predicted Movie Rating: {predicted_rating:.2f}")
```
