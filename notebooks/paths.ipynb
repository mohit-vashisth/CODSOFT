```python
import os

def list_all_file_paths(base_path: str) -> None:
    """
    Recursively print absolute paths of all files inside the given folder.

    Args:
        base_path (str): Path to the folder to scan.
    """
    if not os.path.exists(base_path):
        print(f"‚ùå Path does not exist: {base_path}")
        return

    if not os.path.isdir(base_path):
        print(f"‚ùå Not a directory: {base_path}")
        return

    print(f"\nüìÅ Scanning folder: {os.path.abspath(base_path)}\n")
    
    for root, dirs, files in os.walk(base_path):
        # Show current folder
        print(f"üìÇ Folder: {os.path.abspath(root)}")
        if not files:
            print("   üö´ No files found in this folder.")
        for file in files:
            full_path = os.path.abspath(os.path.join(root, file))
            print(f"   üìÑ {full_path}")
        print()  # spacing between folders

list_all_file_paths(r"C:\Users\Admin\Documents\CODSOFT")
```
